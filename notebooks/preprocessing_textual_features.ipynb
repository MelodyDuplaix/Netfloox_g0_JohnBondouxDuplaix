## Processus pour le Traitement de la colonne actors


Valeurs actuelles de la colonne:


```python
df["actor"].value_counts()
```

Fonction pour nettoyer la colonne (enlever les guillemets autour des valeurs):


```python
def clean_actor_list(actor_list_str):
    if actor_list_str == "[]":
        return []
    if type(actor_list_str) == float:
        return []
    actor_list_str = actor_list_str.strip("[]").replace("'", "").replace('"', "")
    actor_list = actor_list_str.split(", ")
    return [actor for actor in actor_list if actor]
```

Comptage des acteurs:


```python
df["actor"].apply(clean_actor_list).explode().value_counts()
```

On cherche ensuite à encoder


```python
df['actor'].apply(clean_actor_list)
```

Multilabelbinarizer a l'air particulièrement adapté pour des listes de catégories (ici des listes d'acteurs)


```python
mlb = MultiLabelBinarizer()
test = mlb.fit_transform(df["actor"].apply(clean_actor_list))
test
```

```python
mlb.classes_
```

On se retrouve par contre avec un nombre de colonne très conséquent, on peux envisager de réduire les colonnes en ne prenant que les acteurs les plus fréquent (1000 par exemple).


Pour cela on va récupérer la liste des 1000 acteurs les plus fréquent, puis filtrer dans la colonne les acteurs présent dans cette liste pour ne pas garder les autres.


```python
actor_counts = df["actor"].apply(clean_actor_list).explode().value_counts()
top_actors = actor_counts.head(1000).index.tolist()
df["filters_actor"] = df['actor'].apply(clean_actor_list).apply(lambda actor_list: [actor for actor in actor_list if actor in top_actors])
df["filters_actor"]
```

On peux ensuite à nouveau encoder.


```python
mlb = MultiLabelBinarizer()
test = mlb.fit_transform(df["filters_actor"])
test
```

```python
mlb.classes_
```

Transformation finale totale :


```python

def clean_actor_list(actor_list_str):
    if actor_list_str == "[]":
        return []
    if type(actor_list_str) == float:
        return []
    actor_list_str = actor_list_str.strip("[]").replace("'", "").replace('"', "")
    actor_list = actor_list_str.split(", ")
    return [actor for actor in actor_list if actor]

actor_counts = df["actor"].apply(clean_actor_list).explode().value_counts()
top_actors = actor_counts.head(500).index.tolist()
df["filters_actor"] = df['actor'].apply(clean_actor_list).apply(lambda actor_list: [actor for actor in actor_list if actor in top_actors])
mlb_actors = MultiLabelBinarizer()
actors_vector_df = mlb_actors.fit_transform(df["filters_actor"])
actors_encoded_df = pd.DataFrame(actors_vector_df, columns=mlb_actors.classes_)
df_encoded = pd.concat([df.reset_index(drop=True), actors_encoded_df.reset_index(drop=True)], axis=1)
df_encoded
```

Nettoyage de la colonne

On applique ensuite l'ensemble du traitement (nettoyage et vectorisation)


```python
def clean_list(list_str):
    if list_str == "[]":
        return []
    if type(list_str) == float:
        return []
    list_str = list_str.strip("[]").replace("'", "").replace('"', "").replace("\\\\N", "")
    list_str = list_str.split(", ")
    return [region for region in list_str if region]

mlb_region = MultiLabelBinarizer()
region_vector_df = mlb_region.fit_transform(df["regionlist"].apply(clean_list))
region_encoded_df = pd.DataFrame(region_vector_df, columns=mlb_region.classes_)
df_encoded = pd.concat([df_encoded.reset_index(drop=True), region_encoded_df.reset_index(drop=True)], axis=1)
df_encoded
```
